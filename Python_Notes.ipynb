List Methods

1. append(): Adds one element at the end of the list.
2. extend(): Adds multiple elements from another list.
3. insert(): Adds an element at a specific position.
4. remove(): Removes a specified value.
5. pop(): Removes an element using its index.
6. clear(): Removes all elements.
7. index(): Finds the position of an element.
8. count(): Counts how many times an element appears.
9. sort(): Arranges items in ascending order.
10. reverse(): Reverses the list.
11. copy(): Creates a duplicate copy.

*  Functions, not a method

12. len(): Finds the total number of elements.
13. max()/min(): Finds the largest/smallest value.
14. sum(): find the sum of the numbers



```python
# Original List
numbers = [10, 20, 30, 20, 40]

print("Original List:", numbers)

# append()
numbers.append(50)
print("\nAfter append(50):", numbers)

# extend()
numbers.extend([60, 70])
print("After extend([60, 70]):", numbers)

# insert()
# first comes the index then comes the value
numbers.insert(2, 25)
print("After insert(2, 25):", numbers)

# remove()
numbers.remove(20)
print("After remove(20):", numbers)

# pop()
removed = numbers.pop(3)
print("After pop(3):", numbers)
print("Removed Element:", removed)

# index()
print("Index of 30:", numbers.index(30))

# count()
print("Count of 20:", numbers.count(20))

# sort()
numbers.sort()
print("After sort():", numbers)

# reverse()
numbers.reverse()
print("After reverse():", numbers)

# copy()
new_list = numbers.copy()
print("Copied List:", new_list)

# len()
print("Length of List:", len(numbers))

# max()
print("Maximum Value:", max(numbers))

# min()
print("Minimum Value:", min(numbers))

# clear()
new_list.clear()
print("After clear():", new_list)
```

Loops help us:

1. Avoid writing repetitive code

2. Process large amounts of data

3. Work with lists, tuples, strings, and files

4. Perform repeated calculations

Types of Loops in Python
1. for Loop
2. while Loop

1. FOR LOOP

A for loop is used when we know how many times we want to repeat something.



```python
# Syntax
# for variable in sequence:
#     statements
```

range(start, stop, step)

```python
for i in range(2, 11, 2):
    print(i)
```

```python
for i in range(1, 6):
    print(i)
```

```python
num = 5

for i in range(1, 11):
    print(num, "x", i, "=", num*i)
```

```python
Days = ["Monday", "Tuesday", "Wednesday", "Thursday","Friday","Saturday","Sunday"]

for d in Days:
    print(d)
```

```python
name = "Thread Security"

for letter in name:
    print(letter)
```

2. WHILE LOOP: runs as long as a condition remains True.

```python
# Syntax
# while condition:
#     statements
```

```python
i = 1

while i <= 5:
    print(i)
    i += 1
```

```python
count = 5

while count > 0:
    print(count)
    count -= 1
```

Loop Control Statements:
1. break
2. continue
3. pass

1. Break Statement

Stops the loop immediately.

```python
for i in range(1, 11):

    if i == 5:
        break

    print(i)
```

2. Continue Statement

* Skips the current iteration, proceed with the next iteration
* It lets you set back at the initial line of code.


```python
for i in range(1, 6):

    if i == 3:
        continue

    print(i)
```

3. Pass Statement

Placeholder statement.

```python
for i in range(5):

    if i == 3:
        pass

    print(i)
```

Nested For Loop:

For Loop inside a For loop

```python
for i in range (1,7, 2):
    for j in range (i):
        print(i+j)
        if i % 2 == 1:
            break
```

```python
for i in range(3):

    for j in range(4):
        print("*", end=" ")

    print()
```

Functions in Python

A Function is a block of code that performs a specific task and can be reused whenever needed.

In Python, we write a function once and use it many times.

Advantages of Functions:

1. Reusability

2. Reduces Code Duplication

3. Easier Debugging

4. Better Code Organization

5. Improves Readability

Types of Functions
1. Built-in Functions: Functions already provided by Python.

Examples:

1. print()
2. len()
3. max()
4. min()
5. sum()
6. type()

2. User-Defined Functions

Functions created by the programmer.

```python
# Syntax
# def function_name():
#     statements
```

Return Statement

A function can send a value back using return.

```python
def add(a, b):
    return a + b

result = add(10, 20)

print(result)
```

Types of Parametres in Function

| Type                                                   | Description                                                               | Example                     |
| ------------------------------------------------------ | ------------------------------------------------------------------------- | --------------------------- |
| **1. Positional Parameters**                           | Values are passed in the same order as the parameters are defined.        | `add(10, 20)`               |
| **2. Keyword Parameters**                              | Values are passed using the parameter names, so the order doesn't matter. | `add(a=10, b=20)`           |
| **3. Default Parameters**                              | Parameters are assigned a default value if no argument is provided.       | `def greet(name="Student")` |
| **4. Variable-Length Parameters (`*args`)**            | Accepts any number of positional arguments.                               | `def total(*numbers):`      |
| **5. Variable-Length Keyword Parameters (`**kwargs`)** | Accepts any number of keyword arguments as a dictionary.                  | `def student(**details):`   |
